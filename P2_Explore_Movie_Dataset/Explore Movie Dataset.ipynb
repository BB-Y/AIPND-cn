---
**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
df['profit'] = df['revenue'] - df['budget']
profit = df.groupby('release_year').profit.mean()
```

```python
profit.plot(kind='line', title='Profit for Year', color='xkcd:azure', figsize=[15,6]);
```

简要分析：净利润随年份增长增加，增长速度逐渐变低

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
#获取最多产10为导演的名字
director_10 = df.groupby('director').size().nlargest(10)
director_10.reset_index(name='num')
director_10.index

#10位导演所有电影
movie = df.query('director in @director_10.index')
top_3 = movie.groupby('director').revenue.nlargest(3)
top_3.index = top_3.index.droplevel(1)
top_3 = top_3.reset_index(name='rev')
#top_3.drop('level_1', axis=1, inplace = True)
#top_3.plot(kind='bar',x='director');
top_3['num'] = ['a', 'b', 'c'] *10
top_3 = top_3.pivot_table(index='director', values='rev',columns='num')
```

```python
top_3.plot(kind='bar', figsize=[15,5], rot=0);
```

```python
#票房大于0的电影票房均值
df[df['revenue']>0].revenue.mean()
#df.revenue.describe()
```

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
df_6 = df.query('1968<=release_year <= 2015')
df_6.release_year.describe()

df_6 = df_6[df_6.release_date.apply(lambda x: x.month==6)]
df_6.groupby('release_year').id.count().plot();
```

---

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
df_6[df_6.genres.apply(lambda x: 'Comedy' in x.split('|'))].groupby('release_year').id.count().plot(subplots=True, label='Comedy',legend = True);
df_6[df_6.genres.apply(lambda x: 'Drama' in x.split('|'))].groupby('release_year').id.count().plot(subplots=True, label='Drama',color='r',legend = True);
```

#######
1968年~2015年六月电影 Comedy 和 Drama 两类电影的数量基本保持同步增长
